```python
data =  pd.read_csv('./Data/final_new.csv')
data 
```

```python
stations = data.station_id.unique()
```

```python
data['time'] = pd.to_datetime(data['time'])
```

```python
data.isna().sum()
```

```python
data = data.set_index(["time", "station_id"])
```

```python
time = pd.date_range(start='2014-05-01 00:00:00', end='2015-04-30 23:00:00', freq='H')
time
```

```python
data1 = data.reset_index()
```

```python
data_new = pd.DataFrame(index=pd.MultiIndex.from_product([time, stations], names=['time', 'station_id'])).reset_index()
```

```python
data_new
```

```python
data_final = pd.merge(data_new, data, how='left', left_on=['time', 'station_id'], right_on = ['time', 'station_id'])
```

```python
data_final
```

```python
data_final['wind_speed'] = data_final['wind_speed'].interpolate(method="quadratic").ffill().bfill()
data_final['wind_direction'] = data_final['wind_direction'].interpolate(method="nearest").ffill().bfill()
data_final['humidity'] = data_final['humidity'].interpolate(method="quadratic").ffill().bfill()
data_final['pressure'] = data_final['pressure'].interpolate(method="quadratic").ffill().bfill()
data_final['weather'] = data_final['weather'].interpolate(method='nearest').ffill().bfill()
data_final['temperature'] = data_final['temperature'].interpolate(method="quadratic").ffill().bfill()
data_final['PM25_Concentration'] = data_final['temperature'].interpolate(method="quadratic").ffill().bfill()
```

```python
data_final.drop('district_id', axis=1, inplace=True)
```

```python
lat_dic = {}
lon_dic = {}
for i in data1[data1['time']=='2014-05-01 00:00:00'].iterrows():
  lat_dic[i[1]['station_id']] = i[1]['latitude']
  lon_dic[i[1]['station_id']] = i[1]['longitude']
```

```python
for i in stations:
  data_final.loc[data_final['station_id']==i, 'latitude'] = lat_dic[i]
  data_final.loc[data_final['station_id']==i, 'longitude'] = lon_dic[i]
```

```python
data_final
```

```python
data_final.to_csv('Data/data_final_transformer.csv', index=False)
```
